# Intro to the Trees Data
The trees data ranks each neighborhood by how beneficial its trees are (in dollar value) to four different factors:
- Stormwater Benefits
- Property Value
- Electricity Benefits
- Air Quality Value

```python
import pandas as pd
import numpy as np
```

```python
t = pd.read_csv("data/trees_data.csv")
trees = t.iloc[:,[48,3,15,16,17,19,20,37,38,39,47]].dropna().sort_values("neighborhood")
trees.head(10)
```

```python
#How many trees are in each neighborhood?
treesPerNeighborhood = {}

for r in range(len(trees)) :
    neighborhood = trees.iloc[r,0]
    
    try :
        treesPerNeighborhood[neighborhood] += 1
    except :
        treesPerNeighborhood[neighborhood] = 1
        
#for x in treesPerNeighborhood :
#    print(str(x) + " " + str(treesPerNeighborhood[x]))
```

```python
#Isolate each unique neighborhood
n = trees["neighborhood"].unique().tolist()

#change neighborhood column in dataframe to the indicies
for thing in n :
    trees = trees.replace([thing], n.index(thing))
trees.head(10)
```

```python
#Function to collect all of the values in a certain column for each neighborhood.
def statsGrab(c) :
    statList = []
    stat = 0
    index = 0
    for r in range (len(trees)) :
        if (trees.iloc[r,0] == index) :
            stat += float(trees.iloc[r,c])
        else :
            statList.append(stat)
            index+=1
            stat = float(trees.iloc[r,c])
    statList.append(stat)
    return statList

stormValue = statsGrab(2) #Isolate column 2 with storm value dollar benefits
propValue = statsGrab(4) #Isolate column 4 with property value benefits
elecValue = statsGrab(5) #Isolate column 5 with electricity value benefits
airValue = statsGrab(7) #Isolates column 7 with air quality value benefits
```

Now that all the benefits in each category have been totaled for each individual neighborhood, neighborhoods can begin to be ranked. They will be numerically ranked, then all of the ranking scores will be added up. The neighborhoods with the lowest scores will be considered the best!

```python
index = []
for i in range(90) :
    index.append(i)

scoresList = [[n],
              [stormValue],
              [propValue],
              [elecValue],
              [airValue]]
scores = pd.DataFrame(scoresList,
                      columns = ['Neighborhood', 'Stormwater Benefits', 'Property Value', 'Electric Benefits', 'Air Quality Benefits'],
                     index = [index])
```
